**In default,the whiskers extend to 1.5 * the interquartile range**  
1. Make them extend to 2.0 * IQR: whis=2.0  
2. Show the 5th and 95th percentiles: whis=[5,95]  
3. Show min and max values: whis=[0, 100]

```python
#change whiskers
#Omit the outliers
sns.catplot(x='day',y='total_bill',data=tips,kind='box',whis=[5,95])
```

# Add title and label

Seaborn create two different objects: **FacetGrid** and **AxesSubplot**  

**FacetGrid**: relplot(),catplot() - can create subplots
**AxesSubplot**: scatterplot, countplot etc. - only create a single plot

```python
#FacetGrid
sns.set_context('notebook')
g = sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',ci=None)
g.fig.suptitle("Total bills vs. Sex")
```

```python
#adjust height of title
sns.set_context('notebook')
g = sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',ci=None)
g.fig.suptitle("Total bills vs. Sex", y=1.03)
```

```python
#AxesSubplot
c = sns.barplot(x='day',y='total_bill',data=tips)
c.set_title("Total bills",y=1.03)
```

```python
sns.set_context('notebook')
g = sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',col='time',ci=None)
g.fig.suptitle("Total bills vs. Sex", y=1.03)
```

```python
#title for subplots
sns.set_context('notebook')
g = sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',col='time',ci=None)
g.fig.suptitle("Total bills vs. Sex", y=1.03)
g.set_titles('{col_name}')
```

```python
#add axies labels
sns.set_context('notebook')
g = sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',col='time',ci=None)
g.fig.suptitle("Total bills vs. Sex", y=1.03)
g.set_titles('{col_name}')
g.set(xlabel="New X label",ylabel="New Y label")
```

```python
#Rotating x-axis tick labels
#add axies labels
sns.set_context('notebook')
g = sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',ci=None)
g.fig.suptitle("Total bills vs. Sex", y=1.03)
g.set_titles('{col_name}')
g.set(xlabel="New X label",ylabel="New Y label")
plt.xticks(rotation=90)
```
